Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.height = 4, repr.plot.width = 5)
p = ggplot(penguins, aes(x=body_mass_g)) + geom_histogram(na.rm=TRUE, binwidth=100, fill="orange", color="darkorange") +
    facet_wrap(~year, ncol=1) + labs(title = 'Histograms of penguin body mass for measurements taken from 2007-2009',
                                     y = 'Number of penguins', x = 'Body mass in grams') + 
    theme(plot.title.position = 'plot',
          plot.title = element_text(size=10),
          plot.background = element_rect(fill = '#FFFFFF'),
          panel.background = element_rect(fill = '#FFFFFF'),
          strip.background = element_blank(),
          panel.grid = element_line(color = 'grey90'),
          axis.ticks = element_blank(),
          axis.text.x = element_text(size=7))+
    scale_x_continuous(labels = comma_format())
    
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
p = ggplot(penguins, aes(x=body_mass_g, y=sex)) + 
    geom_point(na.rm=TRUE, position = position_jitter(h = 0.25, w = 0.5), size = 1, alpha = 0.5) + 
    facet_wrap(~species, ncol=1)+ 
    labs(y='', x = 'Body mass in grams') + 
    theme(plot.background = element_rect(fill = '#FFFFFF'),
          panel.background = element_rect(fill = '#FFFFFF'),
          strip.background = element_blank(),
          panel.grid = element_line(color = 'grey90'),
          axis.ticks = element_blank(),
          axis.text.x = element_text(size=7))


p
```
